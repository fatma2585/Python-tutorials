Iteration:

Is the ability to run one or more statments mutliple times. This feature is referred to as iterration on looping. 
There are two ways to do this in python using  (For, While loops)
we consider the Iterations as extension of (If statment)?

1. While Loop

while condition:  i<=3

    statment(s)

```python
# Calculating the factorial of a number  
fact=1              
i=1

while i <= 10:                           #As the condition still true the loop continue iterating 
    fact= fact * i
    i=i+1                                
print("factorial of 10 is: {}".format(fact))
```

For Loop:

Is used to iterate the loop over sequences (Lists, tuples, dictionary, strings and ranges)
The statments of the for loop are excuted once for each element in sequence.

for value in sequance:

    statment(s)

```python
#looping over a list or Tuple

days=['Saturday', 'Sun', 'Monday', 'Tuesday', 'wednes', 'Thursday', 'Friday']

for i in days:
    print(i)
```

```python
# Append function
weekends=[]
working_days=[]

for i in range (len(days)):
    if 'Saturday' in days[i]:
        weekends.append(days[i])     
    elif 'Friday' in days[i]:
        weekends.append(days[i])
    else :
        working_days.append(days[i])
        
print("weekends are {}:".format(weekends))
print("working days are {}:".format(working_days))
```

```python
#looping over a string

today ='Monday'

for i in today:
    print(i)
```

```python
#looping over a dictionary
first_year ={'Names':['Salma','Islam','Mohamad',"Nour"] , "Age":17, "subjects":["Chemistry","physics","Astronomy"]}

for i in first_year:
    print("key:", i, "," , "value:", first_year[i])
```

range(n): Creates a sequensce of numers from 0 to n-1

range(a,b):  Creates a sequensce of numers from 0 to b-1

range(a,b,step):  Creates a sequensce of numers from 0 to b-1 and the step value

Break and continue functions 

```python
for i in range(len(days)):
    print("Today is :{}".format(days[i]))
    if (days[i]=="Saturday"):
        print("This day :{} is my offday".format(days[i]))
        break
```

```python
for i in range(len(days)):
    print("Today is :{}".format(days[i]))
    if (days[i]=="Saturday"):
        print("day"+ days[i]+"is my offday")
        continue
    print(days[i]+" is a working day")
```
